```python
llist = {}

offset = 0
while True:
    params = {
        "status": "7",  # all entries
        "offset": offset,
    }
    resp = requests.get("https://myanimelist.net/animelist/sqbl/load.json", params=params)
    if resp.status_code != 200:
        print(resp.status_code)
        break
    print(resp.status_code)
    res = json.loads(resp.text)
    # add to json
    for item in res:
        llist[item['anime_id']] = item['score']
    if len(res) < 300:
        break
    offset += 300
len(llist)
```
